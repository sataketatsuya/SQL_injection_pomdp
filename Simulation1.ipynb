## エピソード数1000回

```python
nepisodes = 10**3

agt = agn.Agent(const.actions,verbose=False)
agt.set_learning_options(exploration=0.02, 
                         learningrate=0.1, 
                         discount=0.9, max_step = 1000)

steps = []; rewards = []; states = []
for _ in tqdm(range(nepisodes)):
    env = SQLenv.mockSQLenv(verbose=False)
    
    agt.reset(env)
    agt.run_episode()
    
    steps.append(agt.steps)
    rewards.append(agt.rewards)  
    states.append(ut.getdictshape(agt.Q)[0])
```

```python
print('学習中の平均報酬：{}'.format(np.mean(np.array(rewards))))
print('学習中の平均ステップ数:{}'.format(np.mean(np.array(steps))))
```

```python
plt.plot(states)
plt.xlabel('episodes')
plt.ylabel('number of states')
```

```python
plt.plot(SGfilter(steps,101,3))
plt.xlabel('episodes')
plt.ylabel('number of steps')
```

```python
plt.plot(np.mean(np.array(rewards).reshape(-1,1),axis=1))
plt.xlabel('episodes')
plt.ylabel('reward')
```

```python
plt.plot(np.mean(np.array(rewards).reshape(-1,10),axis=1))
plt.xlabel('episodes')
plt.ylabel('reward')
```

```python
nepisodes = 100

agt.set_learning_options(exploration=0.0, max_step = 1000)

Tsteps = []; Trewards = []; Tstates = []
for _ in tqdm(range(nepisodes)):
    env = SQLenv.mockSQLenv(verbose=False)
    
    agt.reset(env)
    agt.run_episode()
    
    Tsteps.append(agt.steps)
    Trewards.append(agt.rewards)
    Tstates.append(ut.getdictshape(agt.Q)[0])    
```

```python
print('学習中の平均報酬：{}'.format(np.mean(np.array(rewards))))
print('学習中の平均ステップ数:{}'.format(np.mean(np.array(steps))))
```

```python
print('学習済みの平均報酬：{}'.format(np.mean(np.array(Trewards))))
print('学習済みの平均ステップ数:{}'.format(np.mean(np.array(Tsteps))))
```
